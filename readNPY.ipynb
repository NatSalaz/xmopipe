# Little script to read npy files

```python
import numpy as np
```

```python
file_name = "9534_12_3.npy"
```

**T = Number of frames**

**root**        [T,0:3] = Axis-angle pelvis joint (which is the global orientation of the body)

**body**        [T,3:66] = Axis-angle joints

**hand**        [T,66:156] = Axis-angle joints of hands

**jaw**         [T,156:159] = Axis-angle jaw joint

**expr**        [T,159:209] = 50 dim expressions

**face_shape**  [T,209:309] = 100 face shapes

**trans**       [T,309:312] = Global translation

**betas**       [T,312:322] = Body shape

```python
donnees = np.load(file_name, allow_pickle=True)
print(donnees.shape)
```
