# Little script to read npz files

```python
import numpy as np
```

```python
file_name = "./example.npz"
```

```python
donnees = np.load(file_name, allow_pickle=True)
for key in donnees["body_0"].item().keys():
    if isinstance(donnees["body_0"].item()[key], str) or isinstance(donnees["body_0"].item()[key], int):
        print(key,":", donnees["body_0"].item()[key])
    else: 
        print(key,"shape:", donnees["body_0"].item()[key].shape)
```
