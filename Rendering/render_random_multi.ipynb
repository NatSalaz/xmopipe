```python
import os

os.environ["PYOPENGL_PLATFORM"] = "egl"
import numpy as np
import random
import time
import argparse
import json
# from render_utils.fast_render import render_multi_person_with_overlay
# from render_utils.fast_render import render_multi_person_with_overlay_skeleton
from render_utils.fast_render_CL import (
    render_multi_person_with_overlay,
    render_multi_person_with_overlay_skeleton,
)

root_dir = "../../datasets/videosPPmerged1211/"


def get_num_people(data_npz):
    return len(data_npz.keys())
for i in range (50):
    print(f"Video {i}")
    while True:
        video_dirs = [
            d
            for d in os.listdir(root_dir)
            if d.startswith("video_") and os.path.isdir(os.path.join(root_dir, d))
        ]

        video_dir = random.choice(video_dirs)
        full_video_dir = os.path.join(root_dir, video_dir)

        video_id = video_dir.replace("video_", "")

        npz_files = [f for f in os.listdir(full_video_dir) if f.endswith(".npz")]
        if not npz_files:
            continue

        chosen_npz = random.choice(npz_files)
        npz_path = os.path.join(full_video_dir, chosen_npz)

        scene_number = None
        parts = chosen_npz.replace(".npz", "").split("_")
        if parts[-1].isdigit():
            scene_number = parts[-1]

        if scene_number is None:
            print("NO SCENE NUMBER")

        data = np.load(npz_path, allow_pickle=True)
        num_people = get_num_people(data)

        if num_people == 2:
            print(data.keys())
            break
    json_path = os.path.join(root_dir, video_dir, f"description_videos_{video_dir}.json")

    with open(json_path, "r") as f:
        descriptions = json.load(f)

    scene_number = None
    parts = chosen_npz.replace(".npz", "").split("_")
    for p in parts:
        if p.isdigit():
            scene_number = p
    if scene_number is None:
        scene_number = "0"

    json_key = f"{video_dir}/{scene_number}"
    data_txt = descriptions.get(json_key, "")

    render_multi_person_with_overlay_skeleton(
        npz_file=npz_path,
        output_dir="visus_multi only",
        output_file=f"{video_id}_{scene_number}.mp4",
        text_overlay=data_txt,
    )
    render_multi_person_with_overlay_skeleton(
        npz_file=npz_path,
        output_dir="visus_multi only",
        output_file=f"{video_id}_{scene_number}_above.mp4",
        from_above=True,
        text_overlay=data_txt,
    )
```
