# White pixel remover

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.axis("off")
```

```python
mask = np.all(img == [255, 255, 255], axis=-1)
img_no_white = img.copy()
img_no_white[mask] = [0, 0, 0]
```

```python
plt.imshow(img_no_white)
plt.axis("off")
```

```python
img_rgba = cv2.cvtColor(img, cv2.COLOR_RGB2RGBA)
white = np.all(img == [255, 255, 255], axis=-1)
img_rgba[white, 3] = 0

plt.imshow(img_rgba)
plt.axis("off")
```

```python
plt.imsave("out.png", img_rgba)
```
