# Little script to read npz files

```python
import cv2
import os
import sys
from pathlib import Path
```

```python
video_path = "teaserOverlay2.mp4"  # path to the input video
output_dir = "teaser_frames"  # output directory
stride = int(50)  # extract 1 frame every N frames
```

```python
os.makedirs(output_dir, exist_ok=True)
```

```python
cap = cv2.VideoCapture(video_path)
assert cap.isOpened(), f"Cannot open {video_path}"

frame_id = 0
saved_id = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_id % stride == 0:
        out_path = os.path.join(output_dir, f"{Path(video_path).stem}frame_{saved_id:06d}.jpg")
        cv2.imwrite(out_path, frame)
        saved_id += stride

    frame_id += 1

cap.release()

print(f"{saved_id} frames extracted")
```
