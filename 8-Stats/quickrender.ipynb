```python
from pathlib import Path
from typing import Callable, List, TypeVar

T = TypeVar("T")  # type de la valeur retournée (str, int, float, etc.)


def read_txt_values(
    path: str,
    *,
    skip_empty: bool = True,
    strip_newline: bool = True,
) -> List[T]:

    values: List[T] = []
    with open(path, encoding="utf-8") as fh:
        for line in fh:
            line = line.strip()
            values.append(line)
    return values
```

```python
chosen_names = sorted(read_txt_values("caroline/fichiersadl.txt"))
print(chosen_names)
```

```python
gc.collect()

torch.cuda.empty_cache() 
torch.cuda.ipc_collect()

notebook_dir = os.getcwd()
parent_dir = os.path.abspath(os.path.join(notebook_dir, "../../../../7-Merge"))
sys.path.append(parent_dir)

os.environ["PYOPENGL_PLATFORM"] = "egl"

from render_utils.fast_render_CL import render_multi_person_with_overlay


def vidfile(filename):
    if not filename:
        return None
    match = re.search(r"video_\d+", filename)
    if match:
        return match.group()  # ex: "video_2935"
    return None

i = 0

total = len(chosen_names)

for idx in range(len(chosen_names)):
    filename = sorted(chosen_names)[idx]
    npz_file = f"../../{vidfile(filename)}/{filename}"
    output_dir = f"caroline/"
    render_multi_person_with_overlay(
        npz_file=npz_file,
        output_dir=output_dir,
        output_file=f"{Path(filename).stem}.mp4",
        emotion=False,
        follow_0=True,
    )
    i += 1
    print(i)
```
