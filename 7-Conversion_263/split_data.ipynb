## Split data from our dataset

```python
import pandas as pd
import os
import random
```

```python
seed = 42
input_dir = "./joints"
output_dir = "XmoPipe"

test_proportion = 0.15
val_proportion = 0.05
train_proportion = 0.8

values = [
    f[:-4] for f in os.listdir(input_dir) if os.path.isfile(os.path.join(input_dir, f))
]
random.shuffle(values)
n = len(values)
n_test = int(n * test_proportion)
n_val = int(n * val_proportion)
n_train = n - n_test - n_val

test = values[:n_test]
val = values[n_test : n_test + n_val]
train = values[n_test + n_val :]
train_val = train + val
all_data = values

print(f"Train: {len(train)}")
print(f"Val: {len(val)}")
print(f"Test: {len(test)}")
print(f"All: {len(all_data)}")
```

```python
os.makedirs(output_dir, exist_ok=True)
def write_list(path, data):
    with open(path, "w") as f:
        f.write("\n".join(data))
write_list(os.path.join(output_dir, "train.txt"), train)
write_list(os.path.join(output_dir, "val.txt"), val)
write_list(os.path.join(output_dir, "test.txt"), test)
write_list(os.path.join(output_dir, "train_val.txt"), train_val)
write_list(os.path.join(output_dir, "all.txt"), all_data)

print(f"Files written to {output_dir}/")
```
