```python
import csv
import os
from tqdm import tqdm
import numpy as np
import re
```

```python
paths = []
folders = []
dataset_names = []
for root, dirs, files in os.walk("../5-Merge/videosPPmerged"):
    #     for folder in dirs:
    #         folders.append(os.path.join(root, folder))
    folders.append(root)
    for name in files:
        dataset_name = root.split("/")[2]
        if dataset_name not in dataset_names:
            dataset_names.append(dataset_name)
        if name.endswith(".npz"):
            paths.append(os.path.join(root, name))
print(paths)
```

```python
save_root = "./pose_data"
save_folders = [
    folder.replace("../5-Merge/videosPPmerged", "./pose_data") for folder in folders
]
group_path = [[path for path in paths if name in path] for name in dataset_names]
print(group_path)
dataset_name = paths[0].split('/')[2]

for path in paths:
    save_path = path.replace("../5-Merge/videosPPmerged", "./pose_data")
    save_path = save_path[:-3] + "npy"
    print(save_path)
```

```python
source_paths = []
start_frames = []
end_frames = []
new_names = []


for path in group_path[0]:
    data = np.load(path, allow_pickle=True)
    for key in data:    
        npy_path = path.replace("../5-Merge/videosPPmerged", "./pose_data")
        npy_path = npy_path[:-4] + "_" + key + ".npy"
        source_paths.append(npy_path)
        start_frames.append(data[key].item()["start"])
        end_frames.append(data[key].item()["stop"])
        match = re.search(r"video_(\d+)_merged_scene_(\d+)_body_(\d+)\.npy$", npy_path)
        if match:
            video_id, scene_id, body_id = match.groups()
            new_names.append(f"{video_id}_{scene_id}_{body_id}.npy")
        else:
            print("No correspondance")

print(len(source_paths))
print(source_paths)
print(len(start_frames))
print(start_frames)
print(len(end_frames))
print(end_frames)
print(len(new_names))
print(new_names)
```

```python
combined = sorted(zip(new_names, source_paths, start_frames, end_frames))
new_names_sorted, source_paths_sorted, start_frames_sorted, end_frames_sorted = zip(
    *combined
)

print(len(source_paths_sorted))
print(source_paths_sorted)
print(len(start_frames_sorted))
print(start_frames_sorted)
print(len(end_frames_sorted))
print(end_frames_sorted)
print(len(new_names_sorted))
print(new_names_sorted)
```

```python
csv_path = "index_xmo.csv"

with open(csv_path, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["source_path", "start_frame", "end_frame", "new_name"])
    for src, start, end, new in zip(
        source_paths_sorted, start_frames_sorted, end_frames_sorted, new_names_sorted
    ):
        writer.writerow([src, start, end, new])

print(f"CSV créé : {csv_path}")
```
