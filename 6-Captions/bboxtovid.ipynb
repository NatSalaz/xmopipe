# Script to test the outlines

```python
from pathlib import Path
from bbox_utils import (
    draw_bboxes_to_temp_video,
)
import cv2
import IPython.display as display
from IPython.display import Video
import time
import torch
import re
from fastsam import FastSAM, FastSAMPrompt
```

```python
video_number = 23565
scene_number = 3
model_path="FastSAM-x.pt"
model = FastSAM(model_path)
```

```python
video_number = int(video_number)
scene_number = int(scene_number)

root = Path(".")

video_dir_re = re.compile(rf"video_0*{video_number}$")
video_dirs = [
    d for d in root.iterdir() if d.is_dir() and video_dir_re.fullmatch(d.name)
]
if not video_dirs:
    raise FileNotFoundError("No dossier video found")

video_dir = video_dirs[0]

video_re = re.compile(rf"0*{scene_number}\.mp4$")

video_candidates = [p for p in video_dir.glob("*.mp4") if video_re.fullmatch(p.name)]

if not video_candidates:
    raise FileNotFoundError("No video found")

video_path = video_candidates[0]
npz_re = re.compile(rf"video_0*{video_number}_merged_scene_0*{scene_number}\.npz$")

npz_candidates = [p for p in video_dir.glob("*.npz") if npz_re.fullmatch(p.name)]

if not npz_candidates:
    raise FileNotFoundError("No npz found")

npz_path = npz_candidates[0]

start_time = time.time()
temp_video_path = draw_bboxes_to_temp_video(video_path, npz_path, model=model)

print("Time taken:", round(time.time() - start_time, 2), "s.")
print(f"Tmp video saved at: {temp_video_path[0]}")
```
